```
# violin plot allows me to see how long each country was able to be at the top for proportions of work per population
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
sns.set(rc={'figure.figsize':(8,8)})
sns.set_style('white', {'axes.facecolor':'lightblue'})
ax = sns.violinplot(x="year", y="country", data=top20)

ax.set(xticks=top20.year[1:19])
plt.title('How Long Can the Biggest Fish Swim', fontsize=18)
plt.suptitle('Top Working Proportions per Country', fontsize=10)
plt.show()
```

```
# had to switch my df to a pivot table in order to plot on this matplotlib bar
# ax.patches allows me to loop decimal place to the height as well as format my annotations over each bar
pivot_top20 = top20.pivot(index='year',columns='country', values='emp_per_pop')

colors = ['blue','red','white']
ax = pivot_top20.loc[:,['Qatar','Luxembourg','Cayman Islands']].plot.bar(color=colors, width=1, figsize=(20,7))
for p in ax.patches: 
  ax.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2, p.get_y()+p.get_height()*1.02), ha='center', va='center')
  plt.title('The Labour Bar', fontsize=24, fontweight='bold')
  plt.xlabel('Dominating the Years of the Top 20 Working Pops',fontsize=20,style='italic')
  plt.ylabel('Proportion of Employment by Population',fontsize=20,style='italic')
  plt.xticks(color='black',fontsize=10)
  plt.yticks(color='black',fontsize=10)
  plt.legend(bbox_to_anchor=(-.05,1))
plt.show()
```
